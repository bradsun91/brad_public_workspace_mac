```python
# 先位置后使用reduce铺路：创造一个merge的函数：
def merge_df(df1, df2):
    df1.sort_values('trade_date', inplace = True)
    merged = df1.merge(df2, on = 'trade_date', how = 'outer')
    merged.sort_values('trade_date', inplace = True)
    return merged
```
