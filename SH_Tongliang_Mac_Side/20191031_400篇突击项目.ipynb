## 1. tid

```python
df_new = pd.DataFrame(index = range(0,1000))
```

```python
df_new['tid'] = ""
df_new['title'] = ""
df_new['content'] = ""
df_new['poster'] = ""
df_new['type'] = ""
df_new['fid'] = ""
df_new['first'] = ""
df_new['position'] = ""
```

```python
count = 10
start = 171
for i, index in enumerate(df_new['tid']):
    if count%10!=0:
        df_new['tid'][i] = start
    else:
        start+=1
        df_new['tid'][i] = start
    count+=1
```

```python
df_new
```

```python
df_new['tid'] = df_new["tid"].shift(10)
```

```python
start_reset = 171
df_new['tid'] = df_new['tid'].apply(lambda x: start_reset if str(x)=="nan" else x)
```
