```python
import pandas as pd, numpy as np
```

```python
#Find all sectors in the Chinese Market 
```

```python
list_industry = []
for index,row in data.iterrows():
    if row['industry'] not in list_industry:
        list_industry.append(row['industry'])

list_industry = list(filter(None, list_industry))
print(list_industry)
```

```python
# Find the largest company in each sector based on current 'market capitalization'
```

```python
data2 = pro.daily_basic(ts_code='', trade_date='20200824', fields='ts_code,trade_date,total_mv')
```

```python
print(data2)
```

```python
data2['industry'] = data['industry']
data2['name'] = data['name']
```

```python
print(data2)
```

```python
data2_grouped = data2.groupby('industry')
data2_mv = data2_grouped['total_mv'].idxmax().reset_index()
print(data2_mv)
#data2_mv = data2_mv.reset_index()

#group by industry
#for name in list_industry:
#    print(data2_grouped.get_group(name))
```

```python
for i in range (data2_mv.shape[0]):
    j = int(data2_mv.loc[i, 'total_mv'])
    data2_mv.loc[i,'name'] = data2.loc[j ,'name']
    data2_mv.loc[i,'ts_code']  = data2.loc[j, 'ts_code']
    
# data2_mv gives the companies with maximum market value in each sector
print(data2_mv)
```

```python
#Find the financial indicator for the above companies 
```

```python
all_fin_data = []
for i in range(data2_mv.shape[0]):
    tscode = data2_mv.loc[i,'ts_code']
    df_indicator = pro.fina_indicator(ts_code = tscode)
    all_fin_data.append(df_indicator)
    print('The following is the financial indicator for the company', str(data2_mv.loc[i,'name']))
    print(df_indicator)
```

```python
all_fin_data_df = pd.concat(all_fin_data)
```
