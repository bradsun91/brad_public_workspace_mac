```python
def get_data(code,start,end):
    df = ts.pro_bar(ts_code=code, adj='qfq', start_date=start, end_date=end)
    return df

#获取当前交易日最新的股票代码和简称
def get_code():
    codes = pro.stock_basic(list_status='L').ts_code.values
    return codes

codes = get_code()
codes = list(codes)
```

```python
# Let's see if the codes from data_to_merge_industry is exactly the same as the other code source where we download 
# the price/volume data.

set(data_to_merge_industry['ts_code'])==set(codes)
```

```python
#Find the financial indicator for the above companies/codes/tickers 
```

```python
#### 取全部公司的财务指标数据

all_fin_data = []
count = 1
for tscode in codes:
    if count%200==0:
        print("=======================Sleeping======================")
        time.sleep(60) # Tushare will stop our downloading process unless we decrease our speed
    else:
        print('The following is the financial indicator for the company', count, str(tscode))
        df_indicator = pro.fina_indicator(ts_code = tscode)
        all_fin_data.append(df_indicator)
    count = count+1
#     print(df_indicator)
print("All Finished!")
```

```python
all_fin_data_df_industry = all_fin_data_df.merge(data_to_merge_industry, on = 'ts_code')
all_fin_data_df_industry.drop_duplicates(['end_date','ts_code'], inplace = True)
all_fin_data_df_industry['roe_rank'] = all_fin_data_df_industry.groupby(['end_date','industry'])['roe'].rank(ascending=False)
all_fin_data_df_industry.drop_duplicates('ts_code').groupby('industry').count().sort_values('ts_code').tail(10)
```
