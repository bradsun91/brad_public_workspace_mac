隔夜的黄金价格的涨跌将影响第二天a股市场的涨跌，26天中2天第二天a股市场和隔夜的黄金价格涨跌保持一致。在全球货币中，美元是避险货币，当它经济衰退时候能反映本国经济情况的股票等指数就会下跌，人们担心这样的衰退会牵涉到全球经济，出于避险的考虑人们就把钱换为美元来避险，所以当股票下跌的时候美元得到了支撑。相反，当股票上涨，说明美国经济好转，也许就能使得全球经济好转，人们此时就把美元兑换成其他高息货币来进行套息交易的投资，所以大量的美元被抛售 从而使美元下跌。那么这样就是在一定情况下黄金和股市是同涨同跌。比如，Quantitative easing导致美元流动性泛滥，假设对市场信心有效，可以推高股市，压低美元，推高金价。所以黄金价格和股市正比例变化。

总体来说，从长期看，黄金价格，原油价格和a股市场的涨跌的关心并不稳定。但短期存在正比例变化，隔夜黄金和原油的价格涨跌，很可能代表第二天股票市场的涨跌。所以可以根据短期趋势，可以根据隔夜黄金和原油的价格走势，判断第二天股市的情况
