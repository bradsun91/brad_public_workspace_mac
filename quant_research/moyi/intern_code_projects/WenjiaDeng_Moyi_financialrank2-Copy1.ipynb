```python
import bs4 as bs
import requests
import pandas as pd
import yahoo_fin.stock_info as si

#Scrap tickers from wikipedia and arrange them by sector


html = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')

soup = bs.BeautifulSoup(html.text, features="lxml")

#creating list for each sector
financial=[]
industrials=[]
healthcare=[]
info_tech=[]
communication=[]
discretionary=[]
materials=[]
realestate=[]
utilities=[]
staple=[]
energy=[]

table = soup.find('table', {'class':'wikitable sortable'})

rows = table.findAll('tr')[1:]

for row in rows:
    ticker = row.findAll('td')[0].text
    indus = row.findAll('td')[3].text

    if indus == 'Industrials':
        industrials.append(ticker[:-1])
    if indus == 'Health Care':
        healthcare.append(ticker[:-1])
    if indus == 'Information Technology':
        info_tech.append(ticker[:-1])
    if indus == 'Communication Services':
        communication.append(ticker[:-1])
    if indus == 'Consumer Discretionary':
        discretionary.append(ticker[:-1])
    if indus == 'Consumer Staples':
        staple.append(ticker[:-1])
    if indus == 'Financials':
        financial.append(ticker[:-1])
    if indus == 'Real Estate':
        realestate.append(ticker[:-1])
    if indus == 'Energy':
        energy.append(ticker[:-1])
    if indus == 'Utilities':
        utilities.append(ticker[:-1])
    if indus == 'Materials':
        materials.append(ticker[:-1])
```

```python
def takeSecond(rank):
    return rank[1]
```

For the following code, my intention is to firstly arrange each stock according to their financial metrics, and then to make them into DataFrame for furthur possible manipulaiton.
Therefore, the ranking codes are the same as the last one, and the only difference will be the additional codes used to put data in dataframes.

```python
rank = []
print("There are {} stocks in this sectors".format(len(industrials)))
for stock in industrials:
    print("Getting financial data for:", stock)
    pe = si.get_quote_table(stock)['PE Ratio (TTM)']
    temp=[stock,pe]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

indu=pd.DataFrame()
indu['Stock']=[rank[i][0] for i in range(0,len(rank))]
indu['PE']=[rank[i][1] for i in range(0,len(rank))]

print(indu)
```

```python
rank = []
for stock in discretionary:
    pe = si.get_quote_table(stock)['PE Ratio (TTM)']
    temp=[stock,pe]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

dis=pd.DataFrame()
dis['Stock']=[rank[i][0] for i in range(0,len(rank))]
dis['PE']=[rank[i][1] for i in range(0,len(rank))]

print(dis)
```

```python
rank = []
for stock in materials:
    pe = si.get_quote_table(stock)['PE Ratio (TTM)']
    temp=[stock,pe]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

mat=pd.DataFrame()
mat['Stock']=[rank[i][0] for i in range(0,len(rank))]
mat['PE']=[rank[i][1] for i in range(0,len(rank))]

print(mat)
```

```python
rank=[]
for stock in utilities:
        sheet=si.get_balance_sheet(stock)
        a=sheet.loc['totalLiab'][1]
        b=sheet.loc['totalStockholderEquity'][1]
        de=a/b
        temp=[stock,de]
        rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

ult=pd.DataFrame()
ult['Stock']=[rank[i][0] for i in range(0,len(rank))]
ult['DE']=[rank[i][1] for i in range(0,len(rank))]

print(ult)
```

```python
rank=[]
for stock in realestate:
        sheet=si.get_balance_sheet(stock)
        a=sheet.loc['totalLiab'][1]
        b=sheet.loc['totalStockholderEquity'][1]
        de=a/b
        temp=[stock,de]
        rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

real=pd.DataFrame()
real['Stock']=[rank[i][0] for i in range(0,len(rank))]
real['DE']=[rank[i][1] for i in range(0,len(rank))]

print(real)
```

```python
rank=[]
for stock in info_tech:
    sheet=si.get_income_statement(stock)
    a=sheet.loc["grossProfit"][1]
    b=sheet.loc["totalRevenue"][1]
    grossmargin=a/b
    temp=[stock,grossmargin]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

tech=pd.DataFrame()
tech['Stock']=[rank[i][0] for i in range(0,len(rank))]
tech['Gross Profit Margin']=[rank[i][1] for i in range(0,len(rank))]

print(tech)
```

```python
rank=[]
for stock in healthcare:
    try:
        sheet=si.get_cash_flow(stock)
        a=sheet.loc["totalCashFromOperatingActivities"][1]
        sheet=si.get_balance_sheet(stock)
        b=sheet.loc['totalLiab'][1]
        cashflowcov=a/b
        temp=[stock,cashflowcov]
        rank.append(temp)
    except:
        pass
rank.sort(key=takeSecond,reverse=True)

health=pd.DataFrame()
health['Stock']=[rank[i][0] for i in range(0,len(rank))]
health['Cash Flow Coverage Ratio']=[rank[i][1] for i in range(0,len(rank))]

print(health)

```

```python
rank=[]
for stock in communication:
    sheet=si.get_income_statement(stock)
    a=sheet.loc["grossProfit"][1]
    b=sheet.loc["totalRevenue"][1]
    grossmargin=a/b
    temp=[stock,grossmargin]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

com=pd.DataFrame()
com['Stock']=[rank[i][0] for i in range(0,len(rank))]
com['Gross Profit Margin']=[rank[i][1] for i in range(0,len(rank))]

print(com)
```

```python
rank=[]
for stock in energy:
    sheet=si.get_income_statement(stock)
    a=sheet.loc["ebit"][0]
    sheet=si.get_balance_sheet(stock)
    b=sheet.loc["totalAssets"][1]
    c=sheet.loc["totalCurrentLiabilities"][1]
    roce=a/(b-c)
    temp=[stock,roce]
    rank.append(temp)
rank.sort(key=takeSecond,reverse=True)

ener=pd.DataFrame()
ener['Stock']=[rank[i][0] for i in range(0,len(rank))]
ener['ROCE']=[rank[i][1] for i in range(0,len(rank))]

print(ener)
```

```python
rank=[]
for stock in financial:
    try:
        p=si.get_quote_table(stock)['Quote Price']
        bp=si.get_stats(stock)['Value'][47]
        pb = p/float(bp)
        temp=[stock,float(bp)]
        rank.append(temp)
    except:
        pass

rank.sort(key=takeSecond,reverse=True)

fin=pd.DataFrame()
fin['Stock']=[rank[i][0] for i in range(0,len(rank))]
fin['P/B']=[rank[i][1] for i in range(0,len(rank))]

print(fin)
```

```python
rank=[]
for stock in staple:
    try:
        roe=si.get_stats(stock)['Value'][33]
        temp=[stock,float(roe[:-1])]
        rank.append(temp)
    except:
        pass
rank.sort(key=takeSecond,reverse=True)

stap=pd.DataFrame()
stap['Stock']=[rank[i][0] for i in range(0,len(rank))]
stap['ROE']=[rank[i][1] for i in range(0,len(rank))]

print(stap)
```

```python
#use pd.concat to combine all 11 dataframes into 1 big dartaframe.

frames=[indu,dis,mat,ult,real,tech,health,com,ener,fin,stap]


report=pd.concat(frames,keys=['Industrial','Consumer Discretionary','Material','Utilities','Real Estate','Info Technology''Healthcare','Communication Services','Energy','Financial','Consumer Staple'])

report
```

```python
#export the dataframe to csv, you may want to search for the name of this csv file, because I use mac and I dont know where it will be saved on different systems


report.to_csv('Financialrank.csv')
```
