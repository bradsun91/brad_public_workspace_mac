计算 Hurst Exponent。若其介于0-0.5，序列符合均值回归；若等于0.5，序列为随机游走，若介于0.5-1，序列为趋势跟踪。

```python
def hurst(ts):
    """Returns the Hurst Exponent of the time series vector ts"""
    # Create the range of lag values
    lags = range(5,100)

    # Calculate the array of the variances of the lagged differences
    tau = [np.sqrt(np.std(np.subtract(ts[lag:].tolist(), ts[:-lag].tolist()))) for lag in lags]
   
    # Use a linear fit to estimate the Hurst Exponent
    poly = np.polyfit(np.log(lags), np.log(tau), 1)

    # Return the Hurst exponent from the polyfit output
    return poly[0] * 2.0
```

均值回归策略，返回该条件下的夏普比率

```python
def mean_reversion(close_data, std_range, lookback_window, plot):

    signals = pd.DataFrame(index=close_data.index)
    signals['signal'] = 0.0
    # 计算移动平均和标准差
    signals['mov_avg'] = close_data.rolling(window=lookback_window).mean()
    signals['mov_std'] = close_data.rolling(window=lookback_window).std()
    # 计算 z-score
    signals['mov_z_score'] = (close_data - signals['mov_avg']) / signals['mov_std']

    # 发出信号
    signals['signal'][(lookback_window-1):] = np.where(signals['mov_z_score'][(lookback_window-1):] 
                                            < -std_range , 1.0, 0.0)  
    signals['signal'][(lookback_window-1):].loc[signals['mov_z_score'] > -1.0] = np.where(signals['mov_z_score'][(lookback_window-1):].loc[signals['mov_z_score'] > -1.0]
                                            > std_range , -1.0, 0.0)  

    # 交易
    signals['positions'] = signals['signal']
    signals['positions'].loc[signals['signal'].diff() == 0] = 0 # 不调仓的时候

    change_1 = signals['positions'].loc[signals['positions'] != 0] # 调仓的时候
    change_1 = change_1.loc[change_1.diff() == 0] # 一直在调仓的时候
    signals['positions'][change_1.index] = 0    # 找到一直在调仓的时候，然后把positions col对应的值变成0
    change_2 = signals['positions'].loc[signals['positions'] != 0] #定义不是一直在调仓时的position的序列
    if len(change_2) != 0:
        if signals['positions'][change_2.index[0]] == -1:
            signals['positions'][change_2.index[0]] = 0 
    
    signals['adjusted_signal'] = np.where(signals['positions'].cumsum() == 1 , 1.0, 0.0) 

    if plot:
        # 绘制移动平均和信号图
        fig = plt.figure()

        ax1 = fig.add_subplot(111,  ylabel='Price in $')

        upper_bound = signals['mov_avg'] + std_range * signals['mov_std']
        lower_bound = signals['mov_avg'] - std_range * signals['mov_std']

        # 买入信号
        ax1.plot(signals.loc[signals.positions == 1.0].index, close_data[signals.positions == 1.0],'^', markersize=10, color='m')
         
        # 卖出信号
        ax1.plot(signals.loc[signals.positions == -1.0].index, close_data[signals.positions == -1.0],'v', markersize=10, color='k')
        # 上下界
        upper_bound.plot(ax=ax1, color='steelblue', linestyle='--', lw=2.)
        lower_bound.plot(ax=ax1, color='steelblue', linestyle='--', lw=2.)

        signals['mov_avg'].plot(ax=ax1, color='dimgray', lw=2.)

        close_data.plot(ax=ax1, color='coral', lw=2.)

        # Show the plot
        plt.show()

    # 初始资本
    initial_capital= float(100000.0)
    positions = pd.DataFrame(index=signals.index).fillna(0.0)

    # 买入15股，该数值可调整 
    positions['bit'] = 15*signals['adjusted_signal']
    portfolio = positions.multiply(close_data, axis=0)
    pos_diff = positions.diff()
    # 持有
    portfolio['holdings'] = (positions.multiply(close_data, axis=0)).sum(axis=1)
    #现金
    portfolio['cash'] = initial_capital - (pos_diff.multiply(close_data, axis=0)).sum(axis=1).cumsum()   

    portfolio['total'] = portfolio['cash'] + portfolio['holdings']

    portfolio['returns'] = portfolio['total'].pct_change()

    if plot:
        # 投资组合
        fig = plt.figure()
        ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')
        ax1.plot(portfolio.loc[signals.positions == 1.0].index, portfolio.total[signals.positions == 1.0],'^', markersize=10, color='m')
        ax1.plot(portfolio.loc[signals.positions == -1.0].index, portfolio.total[signals.positions == -1.0],'v', markersize=10, color='k')
        portfolio['total'].plot(ax=ax1, lw=2.)
        plt.show()

    # 策略评估
    returns = (portfolio['total'][-1] - portfolio['total'][0]) / portfolio['total'][0]
    stds = np.std((portfolio['total'] - portfolio['total'][0]) / portfolio['total'][0])
    # 参考国债回报率
    risk_free = 0.1
    if stds == 0:
        sharp = 0
        print(lookback_window)
    else:
        sharp = (returns - risk_free) / stds
    return sharp
```

动量策略，返回该条件下的夏普比率

```python
def momentum(close_data, short_window, long_window, plot) :
    signals = pd.DataFrame(index=close_data.index)
    signals['signal'] = 0.0

    # 计算移动平均
    signals['short_mavg'] = close_data.rolling(window=short_window, min_periods=1, center=False).mean()

    signals['long_mavg'] = close_data.rolling(window=long_window, min_periods=1, center=False).mean()

    # 产生信号
    signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] 
                                            > signals['long_mavg'][short_window:], 1.0, 0.0)   

    #交易
    signals['positions'] = signals['signal'].diff()

    pd.set_option('display.max_rows', None)

    if plot:
        # 绘制移动平均和信号图
        fig = plt.figure()

        ax1 = fig.add_subplot(111,  ylabel='Price in $')

        # 买入信号
        ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='m')
         
        # 卖出信号
        ax1.plot(signals.loc[signals.positions == -1.0].index, signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='k')

        signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)
        
        # Show the plot
        plt.show()
    # 初始资本
    initial_capital= float(100000.0)
    positions = pd.DataFrame(index=signals.index).fillna(0.0)

    # 买入15股，该数值可调整
    positions['bit'] = 15*signals['signal']
    portfolio = positions.multiply(close_data, axis=0)
    pos_diff = positions.diff()
    # 持有
    portfolio['holdings'] = (positions.multiply(close_data, axis=0)).sum(axis=1)
    #现金
    portfolio['cash'] = initial_capital - (pos_diff.multiply(close_data, axis=0)).sum(axis=1).cumsum()   

    portfolio['total'] = portfolio['cash'] + portfolio['holdings']

    portfolio['returns'] = portfolio['total'].pct_change()
    
    if plot:
        # 投资组合
        fig = plt.figure()
        ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')
        ax1.plot(portfolio.loc[signals.positions == 1.0].index, portfolio.total[signals.positions == 1.0],
         '^', markersize=10, color='m')
        ax1.plot(portfolio.loc[signals.positions == -1.0].index, portfolio.total[signals.positions == -1.0],
         'v', markersize=10, color='k')
        portfolio['total'].plot(ax=ax1, lw=2.)

        plt.show()
    
    # 策略评估
    returns = (portfolio['total'][-1] - portfolio['total'][0]) / portfolio['total'][0]
    stds = np.std((portfolio['total'] - portfolio['total'][0]) / portfolio['total'][0])
    risk_free = 0.1
    sharp = (returns - risk_free) / stds
    return sharp
```
