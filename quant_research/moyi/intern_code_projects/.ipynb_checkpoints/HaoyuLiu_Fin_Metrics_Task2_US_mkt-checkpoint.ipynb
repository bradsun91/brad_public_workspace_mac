# Get Top5 Function

    We use GetTop5 Function to select 5 top companies in each sector, based on Market Capitalization. This function return the quote table of the top5 companies.
    
   1. We get tickers from [List of S&P 500 companies](https://en.wikipedia.org/wiki/List_of_S%26P_500_companies)  (Wikipedia)
   2. Since we cannot get quote table and stats data of 'BF.B', 'BRK.B', we directly delete them from the list of tickers.



```python
def getTop5(sector):
    payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    first_table = payload[0]
    second_table = payload[1]

    df=first_table
    df.index=df.Symbol
    df=df.drop(index=['BF.B','BRK.B'])

    sector_df = df[df['GICS Sector'] == sector]
    sector_tickers = sector_df['Symbol'].values.tolist()
    sector_datas=data.get_quote_yahoo(sector_tickers)
    
    sector_datas.sort_values("marketCap",inplace = True, ascending=False)
    Top5=sector_datas.head(5)
    Top5_sector_tickers=Top5.index
    
    return Top5
```

# Rank Top5 Function
    We use Rank_Top5 fuction to rank these 5 top companies in their own industry sector respectively based on 
    their most important financial metrics. This Function return the financial metrics which was used as the 
    basis for ranking. 
    
    1. We use bfill method to fill NaN.
    2. We use Current quater's financial metrics to rank companies, with descending order.
    3. The result cover current quarter's data and 5 quater's historical data.

```python
def Rank_Top5(Top5_tickers, metrics):

    df1= pd.DataFrame()
    #Top5_tickers=Top5.index

    for i in np.arange(0,5,1):
        stats=get_stats_valuation(Top5_tickers[i])
        stats.index=['Market Cap', 'EV', 'Trailing P/E', 'Forward P/E', 'PEG', 'P/S', 'P/B', 'EV/Revenue', 'EV/EBITDA' ]
        data=stats.loc[metrics,:]
        data=pd.DataFrame(data)

        df1=pd.concat([data, df1], axis=1)

    df1.columns=([Top5_tickers[4],Top5_tickers[3],Top5_tickers[2],Top5_tickers[1],Top5_tickers[0]])
    df1=df1.drop(index='Unnamed: 0')
    df1=df1.fillna(method='bfill')
    df1=df1.T
    new_columns=df1.columns.values
    new_columns[0]='Current'
    df1.columns=new_columns
    
    df1=pd.DataFrame(df1,dtype=np.float)
    df1.sort_values(by='Current', inplace=True, ascending=True)
    return df1, data
```
