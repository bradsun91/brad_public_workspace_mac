```python
# sector of each company
# https://datahub.io/core/s-and-p-500-companies#data
```

```python
sector = pd.read_excel('sector.xlsx')  
```

```python
sector_sort = sector.sort_values(['Symbol'])
```

```python
sector_sort
```

```python
df_sort['Sector'] = sector_sort['Sector']
```

```python
df_sort
```

```python
# Q4 top 5 for each sector (rank by weight as company's market capitalization)
df_sort_weight = df_sort.sort_values(['Weight'])
df_top5 = df_sort_weight.groupby('Sector').head(5)
df_top5 = df_top5.sort_values(['Sector', 'Weight'])
```

```python
df_top5
```

# Q5

## 1
For Communication Services industry, as it is hard to find the exact average revenue per user (ARPU), we divide the revenue of first quarter of 2020 by current market capitalization to simulate the ARPU.

```python
df_CommunicationServices = df_top5.loc[df_top5['Sector'] == 'Communication Services']
```

```python
df_CommunicationServices['ARPU']=[1020/0.018479, 2087/0.018995, 232/0.024535, 2683/0.026955, 4425/0.030542]
```

```python
df_CommunicationServices.sort_values(['ARPU'],ascending=False)
```

If a company's P/B ratio is high compared to others in its industry, then the stock might be overvalued. If its P/B ratio is low compared to others in its industry, then it might be undervalued or the company may be performing poorly. Generally, Quanta Services Inc. perform poorer than others.
