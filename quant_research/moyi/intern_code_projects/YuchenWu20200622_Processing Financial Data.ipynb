```python
import pandas as pd
import numpy as np
import requests
import json
import time
from pandas_datareader import data
from yahoo_market_data import Quote
```

```python
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
```

```python
sector = ['Industrials','Health Care','Information Technology','Communication Services','Consumer Discretionary','Utilities','Financials','Materials','Consumer Staples','Real Estate','Energy']
symbol = list(df['Symbol'])
mktcal = []
for i in range(len(symbol)):
    try:
        mktcal.append(data.get_quote_yahoo(symbol[i])['marketCap'][0])
    except:
        mktcal.append(0)
df['mktcal'] = mktcal
df.head(5)
```

```python
def sortindustry(industry):
    industry.sort_values(by = 'mktcal',axis = 0,ascending = False)
    return industry.head(5)
```

```python
def getFinancialMetrics(quote,sector):
    #fr = requests.get(f"https://financialmodelingprep.com/api/v3/key-metrics/{quote}?apikey=62ff5d699f9c24b1196899cb08e2219f")
    #fr = fr.json()
    fr = requests.get(f'https://fmpcloud.io/api/v3/key-metrics/{quote}?apikey=2be675c9d77b598797816f060b897ebe')
    fr = fr.json()
    if sector == 'Industrials' : 
        metrics = fr[0]['inventoryTurnover']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Health Care' : 
        metrics = fr[0]['evToSales']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Information Technology' : 
        metrics = fr[0]['debtToEquity']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Communication Services' : 
        metrics = fr[0]['evToSales']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Consumer Discretionary' : 
        metrics = fr[0]['inventoryTurnover']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Utilities' : 
        metrics = fr[0]['evToSales']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Financials' : 
        metrics = fr[0]['incomeQuality']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Materials' : 
        metrics = fr[0]['evToSales']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Consumer Staples': 
        metrics = fr[0]['inventoryTurnover']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    elif sector == 'Real Estate' : 
        metrics = fr[0]['evToSales']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    else : 
        metrics = fr[0]['inventoryTurnover']
        #metrics = pd.DataFrame(list(metrics.items()),columns=['metrics', quote])
    
    return metrics
```

```python
Industrials = sortindustry(df.loc[df['GICS Sector'] == sector[0]])#固定资产高Operating Cash Flow per Share
x = []
for item in Industrials['Symbol']:
    y = getFinancialMetrics(item,'Industrials')
    x.append(y)
Industrials['inventoryTurnover']  = x
Industrials.sort_values(by = 'inventoryTurnover',axis = 0,ascending = False)
```
