```python
import requests
import pandas as pd
import yfinance as yf
import time
import csv

url = 'https://www.slickcharts.com/sp500'
headers = {"User-Agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
request = requests.get(url, headers = headers)
data = pd.read_html(request.text)[0]
stk_list = data.Symbol
stk_list = data.Symbol.apply(lambda x: x.replace('.', '-'))

print(stk_list[:5])
Technology_list = []
Consumer_Cyclical_list =[]
Communication_Services_list =[]
unknown_list=[]




print('APPL:')
appl_stk_basic_data = yf.Ticker('AAPL').info
print(appl_stk_basic_data['sector'])
appl_marketCap =appl_stk_basic_data['marketCap']
print(appl_marketCap)

print('MSFT:')
msft_stk_basic_data = yf.Ticker('MSFT').info
print(msft_stk_basic_data['sector'])
msft_marketCap =msft_stk_basic_data['marketCap']
print(msft_marketCap)


print('AMZN:')
amzn_stk_basic_data = yf.Ticker('AMZN').info
print(amzn_stk_basic_data['sector'])
amzn_marketCap =amzn_stk_basic_data['marketCap']
print(amzn_marketCap)

print('FB:')
fb_stk_basic_data = yf.Ticker('FB').info
print(fb_stk_basic_data['sector'])
fb_marketCap =fb_stk_basic_data['marketCap']
print(fb_marketCap)

print('GOOG:')
googl_stk_basic_data = yf.Ticker('GOOG').info
print(googl_stk_basic_data['sector'])
googl_marketCap =googl_stk_basic_data['marketCap']
print(googl_marketCap)


#print("Technology: "+sorted_Technology_list)
#print("Consumer Cyclical: "+sorted_Consumer_Cyclical_list)
#print("Communication Services: "+sorted_Communication_Services_list)

for i in stk_list:
    try:
        temp_ticker = yf.Ticker(i).info
        if (temp_ticker['sector'] == 'Technology'):
            temp_info = [temp_ticker['symbol'],temp_ticker['marketCap']]
            Technology_list.append(temp_info)
            print(temp_info)
        elif (temp_ticker['sector'] == 'Consumer Cyclical'):
            temp_info = [temp_ticker['symbol'],temp_ticker['marketCap']]
            Consumer_Cyclical_list.append(temp_info)
            print( temp_info)
        elif (temp_ticker['sector'] == 'Communication Services'):
            temp_info = [temp_ticker['symbol'],temp_ticker['marketCap']]
            Communication_Services_list.append(temp_info)
            print(temp_info)
        time.sleep(1)
    except:
        unknown_list.append(i)
        print(i)
Technology_list.sort(key= lambda Technology_list:Technology_list[1])
Consumer_Cyclical_list.sort(key= lambda onsumer_Cyclical_list:onsumer_Cyclical_list[1])
Communication_Services_list.sort(key= lambda onsumer_Cyclical_list:onsumer_Cyclical_list[1])
print(Technology_list)
with open("Technology.csv", "w", newline="") as A:
    writer = csv.writer(A)
    writer.writerows(Technology_list)
    print("A done")
with open("Consumer_Cyclical.csv", "w", newline="") as B:
    writer = csv.writer(B)
    writer.writerows(Consumer_Cyclical_list)
    print("B done")
with open("Communication_Services.csv", "w", newline="") as C:
    writer = csv.writer(C)
    writer.writerows(Communication_Services_list)
    print("C done")

```
