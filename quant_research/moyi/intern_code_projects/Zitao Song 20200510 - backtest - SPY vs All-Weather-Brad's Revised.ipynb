```python
import matplotlib.pyplot as plt
import pandas_datareader as web
import numpy as np
import pandas as pd
```

```python
symbols = ['SPY', 'TLT', 'DBC', 'SHY', 'GLD']
w = [0.3, 0.4, 0.075, 0.15, 0.075]
base_fund = 1000000
stocks_path = "/Users/miaoyuesun/Code_Workspace/brad_public_workspace_mac/data/US_database/"
```

```python
data_list = []
for symbol in symbols:
    data = pd.read_csv(stocks_path+symbol+".csv")
    data.index = data['Date']
    data_list.append(data)

price_data = pd.concat(data_list, axis=1)
price_data1 = pd.read_csv(stocks_path+"SPY"+".csv")
price_data1.index = price_data1['Date']
price_data = price_data['Adj Close']
price_data1 = price_data1['Adj Close']
```

```python
price_data.dropna(inplace=True)
price_data1 = price_data1[price_data.index]
```

```python
returns_data = price_data.pct_change()[1:]
returns_data1 = price_data1.pct_change()[1:]
weighted_return = (w * returns_data)
port_returns = weighted_return.sum(axis=1)
```

```python
cumulative_returns = (port_returns+1).cumprod()
cumulative_returns1 = (returns_data1+1).cumprod()
returns = (cumulative_returns * base_fund)
returns1 = (cumulative_returns1 * base_fund)
```

```python
returns1.plot(figsize= (18,6))
returns.plot(figsize= (18,6))
```

```python
plt.plot(returns1, label='SPY')
plt.plot(returns, label='All Weather Portfolio')
plt.xlabel('Year')
plt.ylabel('Returns')
plt.title('Simple Backtest')
plt.legend()
```
