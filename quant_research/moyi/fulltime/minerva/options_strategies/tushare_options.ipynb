# 通过Tushare获取上交所期权历史行情数据

https://www.jisilu.cn/?/question/354746

```python
# pip install openpyxl
import openpyxl, time, os.path
```

```python
code = pro.opt_basic(exchange="SSE", fields="ts_code,name")
```

```python
code
```

```python
for index, row in code.iterrows():
    print(index, row['name'])
```

```python
for index, row in code.iterrows():
    name = row['name']#获取期权名称，如”华夏上证50ETF期权1503认购2.30“
    print("正在调取：", name)
    # 如果已经有改名称了，则跳过
    if os.path.exists(name+'.xlsx'):
        pass
    else:
        #根据代码获取所有行情，代码样例“10002234.SH”, 可去上交所查询
        df = pro.opt_daily(ts_code = row['ts_code'], 
                                   fields='ts_code, \
                                           trade_date, \
                                           open,\
                                           high,\
                                           low, \
                                           close,\
                                           vol,\
                                           amount,\
                                           oi')
        wb = openpyxl.Workbook()
        ws = wb.active
        ws.title = row['name']
        data_excel = []
        for index, row in df.iterrows():
            data_excel.append([row['ts_code'],
                               row['trade_date'],
                               row['open'],
                               row['high'],
                               row['low'],
                               row['close'],
                               row['vol'],
                               row['amount'],
                               row['oi']])
        for i in data_excel:
            ws.append(i)
        #excel名称用的是“华夏上证50ETF期权1503认购2.3”
        wb.save(name+'.xlsx')
        data_excel = []
        time.sleep(0.6)
```
