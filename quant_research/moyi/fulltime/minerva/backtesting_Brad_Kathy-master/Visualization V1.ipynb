```python
import pandas_datareader.data as web

sp500 = web.DataReader('^GSPC', 'yahoo', '2012-03-01', '2020-01-09')['Adj Close']


df['ret_sp500'] = sp500.pct_change()

temp = df[['returns','ret_sp500']]

comp = (temp+1).cumprod()
```
